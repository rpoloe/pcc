```python
%%capture
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from google.colab import files
```

```python
files.upload()
data = pd.read_csv("data-pcc.csv", sep=";")
dataset = data.values
np.take(dataset,np.random.permutation(dataset.shape[0]),axis=0,out=dataset)
total_size = len(dataset)
train_size = round(total_size*0.7)
test_size = round(total_size*0.2)
learning_size = train_size + test_size
validation_size = round(total_size*0.1)
train_dataset = dataset[0:train_size]
test_dataset = dataset[train_size:learning_size]
validation_dataset = dataset[learning_size:total_size]
input_train, output_train = train_dataset[:,0:10], train_dataset[: ,10:14]
input_test, output_test = test_dataset[:,0:10], test_dataset[: ,10:14]
input_validation, output_validation = validation_dataset[:,0:10], validation_dataset[: ,10:14]
```

```python
def build_model():
  model = keras.Sequential([
    layers.Dense(800, activation="tanh", input_shape=[10]),
    layers.Dense(640, activation="tanh"),
    layers.Dense(160, activation="tanh"),
    layers.Dense(40, activation="tanh"),
    layers.Dense(4)])
  optimizer = tf.keras.optimizers.Adamax(learning_rate=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-07, name="Adamax")
  model.compile(loss="mse", optimizer=optimizer, metrics=["mae", "mape", "msle", "logcosh"])
  return model
model = build_model()
epochs = 500
history = model.fit(input_train, output_train, epochs=epochs, validation_data=(input_test, output_test), verbose=0)
model.summary()
```
